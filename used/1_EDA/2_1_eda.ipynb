```python
import pandas as pd
import matplotlib.pyplot as plt 
plt.rc('font',family='AppleGothic')
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')

df_raw=pd.read_csv('../data/pps/train_data.csv',index_col=0)
df=df_raw.copy()
df.sort_index(inplace=True)

df=df[df['월'].isin(range(12))]
df_raw=df_raw[df_raw['월'].isin(range(12))]

sample_submission_raw=pd.read_csv('../data/pps/test_data.csv',index_col=0)
sample_submission=sample_submission_raw.copy()

df.index = pd.to_datetime(df.index,format='%Y-%m-%d')
df_raw.index = pd.to_datetime(df_raw.index,format='%Y-%m-%d')
sample_submission.index = pd.to_datetime(sample_submission.index,format='%Y-%m-%d')
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
# for col in df.columns[:4]:
#     df[col] = smooth(df[col],50)
```

```python
df_2018 = df[df.index.year == 2018]
df_2018_raw = df_raw[df_raw.index.year == 2018]
df_2019 = df[df.index.year == 2019]
df_2019_raw = df_raw[df_raw.index.year == 2019]
df_2020 = df[df.index.year == 2020]
df_2020_raw = df_raw[df_raw.index.year == 2020]
df_2021 = df[df.index.year == 2021]
df_2021_raw = df_raw[df_raw.index.year == 2021]
```

```python
for col in df_2018.columns[:4]:
    df_2018[col][10:-10] = smooth(df_2018[col],20)[10:-10]
for col in df_2019.columns[:4]:
    df_2019[col][10:-10] = smooth(df_2019[col],20)[10:-10]
for col in df_2020.columns[:4]:
    df_2020[col][10:-10] = smooth(df_2020[col],20)[10:-10]
for col in df_2021.columns[:4]:
    df_2021[col][10:-10] = smooth(df_2021[col],20)[10:-10]
```

```python
import seaborn as sns
plt.figure(figsize=(14,10))
plt.subplot(2,2,1)
sns.lineplot(data=df_2018.reset_index(),
             x=range(len(df_2018)),
             y='광진구')
sns.lineplot(data=df_2018_raw.reset_index(),
             x=range(len(df_2018_raw)),
             y='광진구')

plt.subplot(2,2,2)
sns.lineplot(data=df_2018.reset_index(),
             x=range(len(df_2018)),
             y='동대문구')
sns.lineplot(data=df_2018_raw.reset_index(),
             x=range(len(df_2018_raw)),
             y='동대문구')

plt.subplot(2,2,3)
sns.lineplot(data=df_2018.reset_index(),
             x=range(len(df_2018)),
             y='성동구')
sns.lineplot(data=df_2018_raw.reset_index(),
             x=range(len(df_2018_raw)),
             y='성동구')

plt.subplot(2,2,4)
sns.lineplot(data=df_2018.reset_index(),
             x=range(len(df_2018)),
             y='중랑구')
sns.lineplot(data=df_2018_raw.reset_index(),
             x=range(len(df_2018_raw)),
             y='중랑구')
plt.show()
```
